## 3. Fonctions élémentaires
### 3.1. Construction d'un RBM et test sur Binary AlphaDigits

```python
X = lire_alpha_digit("2", "A")
image_size = (20, 16)
plot_grid(X, image_size=image_size)
```

```python
X = lire_alpha_digit("A")
plot_grid(X, image_size=image_size)
```

```python
hparams = {
    "p": 320,
    "q": 50,
    "n_epochs": 500,
    "learning_rate": 0.1,
    "batch_size": 10,
}

rbm = RBM(hparams["p"], hparams["q"])

rbm.train(
    X,
    n_epochs=hparams["n_epochs"],
    learning_rate=hparams["learning_rate"],
    batch_size=hparams["batch_size"],
)
```

```python
Y = rbm.generer_image(20, 10)
plot_grid(Y, image_size=image_size)
```

### 3.2. Construction d'un DBN et test sur Binary AlphaDigits

```python
hparams = {
    "config": [(320, 200), (200, 100), (100, 50)],
    "n_epochs": 500,
    "learning_rate": 0.1,
    "batch_size": 10,
}

dbn = DBN(hparams["config"])

dbn.train(
    X,
    n_epochs=hparams["n_epochs"],
    learning_rate=hparams["learning_rate"],
    batch_size=hparams["batch_size"],
)
```

```python
Y = dbn.generer_image(20, 10)
plot_grid(Y, image_size=image_size)
```

### 3.3. Construction d'un DNN et test sur MNIST

```python
X_train, y_train = lire_MNIST("train")
image_size = (28, 28)
plot_grid(X_train, image_size=image_size)
```

```python
hparams = {
    "config": [(784, 200), (200, 100), (100, 10)],
    "n_epochs": 100,
    "learning_rate": 0.1,
    "batch_size": 60000,
}

dnn = DNN(hparams["config"])
```

```python
dnn.retropropagation(
    X_train,
    y_train,
    n_epochs=hparams["n_epochs"],
    learning_rate=hparams["learning_rate"],
    batch_size=hparams["batch_size"],
)
```

```python
hparams = {
    "config": [(784, 200), (200, 100), (100, 10)],
    "n_epochs": 100,
    "learning_rate": 0.1,
    "batch_size": 10000,
}

dnn = DNN(hparams["config"])
```

```python
dnn.retropropagation(
    X_train,
    y_train,
    n_epochs=hparams["n_epochs"],
    learning_rate=hparams["learning_rate"],
    batch_size=hparams["batch_size"],
)
```

```python
hparams = {
    "config": [(784, 200), (200, 100), (100, 10)],
    "n_epochs": 100,
    "learning_rate": 0.01,
    "batch_size": 1000,
}

dnn = DNN(hparams["config"])
```

```python
dnn.retropropagation(
    X_train,
    y_train,
    n_epochs=hparams["n_epochs"],
    learning_rate=hparams["learning_rate"],
    batch_size=hparams["batch_size"],
)
```

```python
X_test, y_test = lire_MNIST("test")
```

```python
dnn.test(X_test, y_test)
```
